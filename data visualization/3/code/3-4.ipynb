```python
import pandas as pd
# 读取数据
data = pd.read_csv('../data/employee.csv')
print(data)
```

```python
import matplotlib.pyplot as plt
# 设置中文字体
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.rcParams['axes.unicode_minus'] = False
name = data.columns #提取数据标签
values = data.values # 提取数据值
plt.figure(figsize=(9,7))
# 绘制散点图
plt.xlabel('年份')
plt.ylabel('各产业就业人员(万人)')
plt.xticks(range(0,20),values[range(0,20),0])
plt.scatter(values[:,0],values[:,2],marker='o',c='red')
plt.scatter(values[:,0],values[:,3],marker='D',c='green')
plt.scatter(values[:,0],values[:,4],marker='v',c='purple')
plt.title('2000-2019年各产业就业人员散点图')
plt.legend([name[2],name[3],name[4]])
plt.savefig('../tmp/2000-2019年各个产业人员散点图.png')
plt.show()
```

```python
# 绘制折线图
plt.figure(figsize=(9,7))
plt.xlabel('年份')
plt.ylabel('产业就业人数(万人)')
plt.xticks(range(0,20),values[range(0,20),0])
plt.title('2000-2019年各产业人员折线图')
plt.plot(values[:,0],values[:,2],c='red',linestyle='--')
plt.plot(values[:,0],values[:,3],c='green',linestyle='-.')
plt.plot(values[:,0],values[:,4],c='blue',linestyle='-')
plt.legend([name[2],name[3],name[4]])
plt.savefig('../tmp/2000__2019年各个产业人员折线图.png')
plt.show()

```

```python
plt.figure(figsize=(7,7))
explode=[0.01,0.01,0.01]
label = name[2:5]
plt.pie(values[-1,2:5],explode=explode,labels=label,autopct='%1.2f%%')
plt.title('2019年各产业就业人员饼图')
plt.savefig('../tmp/2019年各产业人员就业饼图.png')
plt.show()
```

```python
label=name[2:5]
plt.figure(figsize=(9,7))
plt.xlabel('年份')
plt.ylabel('产业就业人数')
plt.xticks(range(3),label)
plt.bar(range(3),values[-1,2:5],width=0.5)
plt.title('2019年各产业就业人员柱状图')
plt.savefig('../tmp/2019年各产业就业人员柱状图.png')
plt.show()
```

```python
label=name[2:5]
data1 = (list(values[:,2]),list(values[:,3]),list(values[:,4]))
plt.figure(figsize=(8,7))
plt.boxplot(data1,labels=label,notch=True,meanline=True)
plt.title('2000-2019年各产业就业人数箱线图')
plt.savefig('../tmp/2000-2019年各产业就业人数箱线图.png')
plt.show()
```
