```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = 'SimHei' # 设置中文字体
plt.rcParams['axes.unicode_minus'] = False
data = pd.read_csv('../data/people.csv')
name = data.columns # 提取其中的columns字段,视为数据的标签
values = data.values # 提取其中的values字段,视为数据的存在位置
plt.figure(figsize=(9,7)) # 设置画布
plt.xlabel('年份') # 添加x轴标签
plt.ylabel('年末总人口(万人)') # 添加y轴标签
plt.xticks(range(0,20),values[range(0,20),0],rotation=45)
plt.title('2000-2019年年末总人口散点图') # 添加图表标题
plt.scatter(values[:,0],values[:,1],c='red',marker='o') # 绘制散点图
plt.savefig('../tmp/2000-2019年年末总人口散点图.png')
plt.show()
print(data)
```

```python
plt.figure(figsize=(9,7))
plt.xlabel('年份') # 添加x轴标签
plt.ylabel('年末总人口(万人)') # 添加y轴标签
plt.xticks(range(0,20),values[range(0,20),0],rotation=45)
plt.title('2000-2019年各个年龄段年末总人口散点图') # 添加图表标题
plt.scatter(values[:,0],values[:,2],marker='o') # 绘制散点图
plt.scatter(values[:,0],values[:,3]) # 绘制散点图
plt.scatter(values[:,0],values[:,4]) # 绘制散点图
plt.legend(['0-14岁人口','15-64岁人口','65及以上人口']) # 添加图例
plt.savefig('../tmp/2000-2019年各个年龄段年末总人口散点图.png')
plt.show()
```

```python
plt.figure(figsize=(9,7))
plt.xlabel('年份') # 横轴标签
plt.ylabel('年末总人口(万人)') #纵轴标签
plt.xticks(range(0,20),values[range(0,20),0],rotation=45)
plt.title('2000-2019年年末总人口折线图') # 添加图标标题
plt.plot(values[:,0],values[:,1],color='r',linestyle='--',marker='o')
plt.savefig('../tmp/2000-2019年年末总人口点线图.png')
plt.show()
```
