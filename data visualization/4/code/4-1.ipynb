```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

# 设置中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']
sns.set_style({'font.sans-serif':['simhei','Arial']})

# 加载数据
hr = pd.read_csv('../data/hr.csv',encoding='gbk')
# 取数据前100条
data = hr.head(100)

# 使用matplotlib库绘图
plt.xlabel('平均每个月工作时长(小时)') # x轴标签
plt.ylabel('满意度水平') #y轴标签
plt.title('满意度水平与平均每个月工作小时') #图表标题
color_map = dict(zip(data['薪资'].unique(),['b','y','r']))
for species, group in data.groupby('薪资'):
    plt.scatter(group['每月平均工作小时数（小时）'],
               group['满意度'],
               color=color_map[species],alpha=0.4,
               edgecolors=None,label=species)
plt.legend(frameon=True,title='薪资')
plt.show()
```

```python
# 使用seaborn绘图
sns.lmplot(x='每月平均工作小时数（小时）',y='满意度',data=data,hue='薪资',fit_reg=False,height=4)
plt.xlabel('平均每个月工作时长(小时)')
plt.ylabel('满意度水平')
plt.title('满意度水平与平均每个月工作小时')
plt.show()
```

```python
# 偏移直线图像展示绘图风格
plt.rcParams['axes.unicode_minus'] = False
x = np.arange(1,10,2)
y1 = x + 1
y2 = x + 3
y3 = x + 5
# 绘制三条不同的直线
# 第一部分,matplotlib的绘图风格
plt.title('Matplotlib的绘图风格')
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
# 第二部分,使用seaborn绘图
sns.set_style('darkgrid') #h设置绘图风格(全黑风格)
sns.set_style({'font.sans-serif':['simhei','Arial']})
sns.lineplot(x=x,y=y1)
sns.lineplot(x=x,y=y2)
sns.lineplot(x=x,y=y3)
plt.title('seaborn库的绘图风格')
plt.show()
```

```python
# 使用偏移直线展示各主题以及修改默认参数示例
# 数据
x = np.arange(1,10,2)
y1 = x + 1
y2 = x + 3
y3 = x + 5
# 绘图函数
def showLine(flip=1):
    sns.lineplot(x=x,y=y1)
    sns.lineplot(x=x,y=y2)
    sns.lineplot(x=x,y=y3)
pic = plt.figure(figsize=(12,8))
#使用darkgrid主题
with sns.axes_style('darkgrid'):
    pic.add_subplot(2,3,1)
    showLine()
    plt.title('darkgrid')
#使用whitegrid主题
with sns.axes_style('whitegrid'):
    pic.add_subplot(2,3,2)
    showLine()
    plt.title('whitegrid')
# 使用dark主题
with sns.axes_style('dark'):
    pic.add_subplot(2,3,3)
    showLine()
    plt.title('dark')
# 使用white主题
with sns.axes_style('white'):
    pic.add_subplot(2,3,4)
    showLine()
    plt.title('white')
# 使用ticks主题
with sns.axes_style('ticks'):
    pic.add_subplot(2,3,5)
    showLine()
    plt.title('ticks')
sns.set_style(style='darkgrid',rc={'font.sans-serif':['MicrosoftYaHei','SimHei'],
                                   'grid.color':'black'}) # 修改主题中的参数
pic.add_subplot(2,3,6)
showLine()
plt.title('修改参数')
plt.show()
```

```python
# 使用四种不同缩放类型绘制图形
sns.set()
# 数据
x = np.arange(1,10,2)
y1 = x + 1
y2 = x + 3
y3 = x + 5
# 制作绘图函数
def showLine(flip=1):
    sns.lineplot(x=x,y=y1)
    sns.lineplot(x=x,y=y2)
    sns.lineplot(x=x,y=y3)
pic = plt.figure(figsize=(8,8),dpi=100)
# 恢复默认参数
# 选择paper类型
with sns.plotting_context('paper'):
    pic.add_subplot(2,2,1)
    showLine()
    plt.title('paper')
# 选择notebook类型
with sns.plotting_context('notebook'):
    pic.add_subplot(2,2,2)
    showLine()
    plt.title('notebook')
with sns.plotting_context('talk'):
    pic.add_subplot(2,2,3)
    showLine()
    plt.title('talk')
with sns.plotting_context('poster'):
    pic.add_subplot(2,2,4)
    showLine()
    plt.title('poster')
plt.show()

```

```python
# 控制图形边框
# 设置中文字体
sns.set_style({'font.sans-serif':['SimHei','Arial']})
# 绘制图形
with sns.axes_style('white'):
    showLine()
#     默认为无参数状态,就删除顶部和右侧的边框
    sns.despine()
    plt.title('控制图形边框')
plt.show()
```
