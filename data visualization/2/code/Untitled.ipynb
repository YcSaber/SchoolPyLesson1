```python
import pandas as pd
data1 = pd.read_csv('../data/数据1.csv',encoding='gb18030',low_memory=False)
data2 = pd.read_csv('../data/数据2.csv',encoding='gb18030',low_memory=False)
print(data1)
print(data2)
```

```python
data1_1 = pd.merge(data1,data2,on='商品')
print(data1_1)
data1_1.to_csv('../data/data1-1.csv',sep=';',index=False)
```

```python
def delRep(list1):
    list2 = []
    for i in list1:
        if i not in list2:
            list2.append(i)
    return list2
data2_1 = list(data1_1['订单号'])
print(len(data2_1))
data2_1 = delRep(data2_1)
print(len(data2_1))
```

```python
data2_2 = set(data2_1)
print(len(data2_2))
```

```python
data2_3 = data1_1['订单号'].drop_duplicates()
print(data2_3)
```
