```python
import matplotlib.pyplot as plt
from sklearn import datasets
import pandas as pd
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=['SpealLength','Spealwidth','PetalLength','Petalwidth'])
df.columns = ['花萼长度','花萼宽度','花瓣长度','花瓣宽度']
df.head()
```

```python
# Kernel Density Estimate(KDE)图也被称为密度图、核密度估计图
df.plot(kind='kde')
plt.ylabel('核密度')# 添加y轴名称
plt.show()
```

```python
df.plot(x='花萼长度', y='花萼宽度', kind='scatter')
plt.show()
```

```python
from sklearn.datasets import load_iris
import seaborn as sns
iris = load_iris()
d = pd.DataFrame(iris.data,columns=['sepal_length','sepal_width','petal_length','petal_width'])
d.columns = ['花萼长度','花萼宽度','花瓣长度','花瓣宽度']
d['species'] = iris.target # 增加一列,表示鸢尾花的类别
sns.countplot(x='species',data=d)
plt.xlabel('类别') # 添加x轴标签
plt.ylabel('数量(个)') # 添加y轴名称
plt.show()
```
